### Preparacion de los Datos: Ventanas Deslizantes

```python
def create_dataset(X, y, time_steps=1): # time_stpes: tamaño de la ventana
    # crea dos listas vacias para depositar los datos
    Xs, ys = [], []
    # el primer lote de datos empieza en la primera observación
    # y toma time_steps  datos.
    # Comienza a avanzar hacia adelante.
    for i in range(len(X) - time_steps):
        v = X.iloc[i:(i + time_steps)].values
        Xs.append(v)
        ys.append(y.iloc[i + time_steps])
    return np.array(Xs), np.array(ys)
```

```python
# Tamaño de la ventana: Numero de rezagos considerados
time_steps = 30

# reshape to [samples, time_steps, n_features]
X_train, y_train = create_dataset(df2[0:train_size], df2[0:train_size], time_steps)
X_test, y_test = create_dataset(df2[train_size:], df2[train_size:], time_steps)

print("Dimension de Entrada:", X_train.shape)
print("Dimension de Salida:", y_train.shape)
```

### Usanod la funcion multivariate_data

```python
def multivariate_data(dataset, target, start_index, end_index, history_size,
                      target_size, step, single_step=False):
    ''' dataset: conjunto de datos para las secuencias de entrada
        target:  conjunto de datos para las secuencias de salida
        start_index: índice inicial de donde empezar a tomar los datos
        end_index: índice final para tomar los datos. None para tomarlos todos
        history_size: tamaño de la venytana para crear las secuencias
        target_size: dentro de cuántas observaciones futuras desea pronosticar
        step: Cada cuántas observaciones deben tomarse. Para secuencias muy largas
              que no cambia mucho localmente
        single_step: Predecir solamente un valor futuro (=True),
                     o predecir todos los valores hasta target_size(=False)
    '''  
    data = []
    labels = []

    start_index = start_index + history_size
    if end_index is None:
        end_index = len(dataset) - target_size

    for i in range(start_index, end_index):
        indices = range(i-history_size, i, step)
        data.append(dataset[indices])

        if single_step:
            labels.append(target[i+target_size])
        else:
            labels.append(target[i:i+target_size])

    return np.array(data), np.array(labels)
```

```python
len_train = 1292
history_size = 30
future_target = 5
STEP = 1

X_train, y_train = multivariate_data(data_scale, data_scale, 0,
                                     len_train, history_size,
                                     future_target, STEP)

X_test, y_test = multivariate_data(data_scale, data_scale, len_train,
                                   None, history_size,
                                   future_target, STEP)
```

### Creacion de tf.data.Dataset
Crea objetos tf.data.Dataset para entreno y validación

```python
BATCH_SIZE = 256
BUFFER_SIZE = 10000

train_data_multi = tf.data.Dataset.from_tensor_slices((X_train, y_train))
train_data_multi = train_data_multi.cache().shuffle(BUFFER_SIZE).batch(BATCH_SIZE).repeat()

val_data_multi = tf.data.Dataset.from_tensor_slices((X_test, y_test))
val_data_multi = val_data_multi.batch(BATCH_SIZE).repeat()
```

### Ejemplo de un Batch

```python
for i in train_data_multi.take(1):
    count = 0
    if count > 0: break
    print(i)
    count += 1
```

### Creacion del Modelo LSTM

```python
X_train.shape
```

```python
multi_step_model = tf.keras.models.Sequential()
multi_step_model.add(tf.keras.layers.LSTM(64, input_shape = X_train.shape[1:]))
multi_step_model.add(tf.keras.layers.Dense())

multi_step_model.compile(optimizer=Adam(0.001), loss='mse')
```

### Summary del Modelo

```python
multi_step_model.summary()
```

### Compilar el Modelo

```python
# Entrenar el modelo
EVALUATION_INTERVAL = 200
multi_step_model.fit(train_data_multi, epochs=10,
                     steps_per_epoch=EVALUATION_INTERVAL,
                     validation_data=val_data_multi,
                     validation_steps=50)
```
